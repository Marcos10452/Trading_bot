```python
import yfinance as yf
#import plotly.graph_objects as go
import pandas as pd
```

```python
def get_stocks(tickers_str, start_date, end_date, interval):
    data=pd.DataFrame()
    data = yf.download(
        tickers = tickers_str,
        start = start_date,
        end = end_date,
        interval = interval,
        actions=True,
        group_by = 'ticker',
        auto_adjust = False,
        prepost = True,
        threads = True,
        proxy = None
        )
    return data
```
